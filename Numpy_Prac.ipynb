```python
## Iterating over array
```

```python
a=np.arange(0,45,5)
a.reshape(3,3)
```

```python
for x in np.nditer(a):
  print(x)
```

```python
# Iteration order( c-style, f-style)
```

```python
for x in np.nditer(a,order="C"):
  print(x)
```

```python
for x in np.nditer(a,order="F"):
  print(x)
```
